```python
import numpy as np
import matplotlib.pyplot as plt
# from utils import *
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
```

```python
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names
target_names = iris.target_names

classes = np.unique(y)

print(feature_names)
print(target_names)
```

```python
def plot_hist(iris):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.ravel()
    for idx, ax in enumerate(axes):
        ax.hist(iris.data[:, idx], bins=20, color='blue', alpha=0.7)
        ax.set_title(iris.feature_names[idx])
    plt.tight_layout()
    plt.show()

plot_hist(iris)
```

```python
def plot_kde(iris):
    df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
    df['species'] = iris.target
    df['species'] = df['species'].apply(lambda x: iris.target_names[x])
    
    for feature in iris.feature_names:
        plt.figure(figsize=(8, 4))
        sns.kdeplot(data=df, x=feature, hue='species', fill=True)
        plt.title(f'KDE Plot of {feature}')
        plt.show()

plot_kde(iris)
```

```python
def plot_scatter(iris):
    df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
    df['species'] = iris.target
    df['species'] = df['species'].apply(lambda x: iris.target_names[x])
    
    sns.pairplot(df, hue='species', markers=["o", "s", "D"])
    plt.show()

plot_scatter(iris)
```

```python
centroids = np.array(
    [np.mean(X[y == i], axis=0) for i in classes]
)

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50, alpha=0.5)
ax.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x')
plt.show()
```
